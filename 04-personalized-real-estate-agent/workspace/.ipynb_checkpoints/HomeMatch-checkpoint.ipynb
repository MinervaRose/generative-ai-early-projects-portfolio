## Step 2: Generate real estate listings using an LLM
We'll use OpenAI (via LangChain) to generate at least 10 diverse and realistic listings. Each listing will include structured info like:

- Neighborhood
- Price
- Bedrooms
- Bathrooms
- House Size
- Description
- Neighborhood Description

We'll define a prompt and run it in a loop to generate 10 listings.

```python
# Initialize LLM
llm = OpenAI(model_name="gpt-3.5-turbo-instruct", temperature=0.7)

# Prompt Template
prompt_template = """
Generate a realistic real estate listing in the following format:

Neighborhood:
Price:
Bedrooms:
Bathrooms:
House Size:
Description:
Neighborhood Description:

Make sure the listing is diverse from previous ones and includes unique features, but remains realistic.
"""

# Generate multiple listings
def generate_listings(num_listings=10):
    listings = []
    for i in range(num_listings):
        print(f"Generating listing {i+1}...")
        response = llm(prompt_template)
        listings.append(response.strip())
    return listings

# Run the generator
raw_listings = generate_listings(10)
```

```python
# Preview the first listing
print(raw_listings[0])
```

```python
# Save to file for later use
with open("listings.txt", "w") as f:
    for listing in raw_listings:
        f.write(listing + "\n\n---\n\n")
```

#### Parse listings into structured format
We’ll turn the GPT-generated text listings into Python dictionaries, one for each listing. This will allow us to:
- Convert each listing into a vector later (Step 3).
- Keep all fields organized for display, personalization, and debugging.

```python
import re

# Helper function to parse a raw listing into a structured dictionary
def parse_listing(text):
    fields = {
        "Neighborhood": "",
        "Price": "",
        "Bedrooms": "",
        "Bathrooms": "",
        "House Size": "",
        "Description": "",
        "Neighborhood Description": ""
    }

    # Regex pattern for fields
    for key in fields:
        pattern = rf"{key}:(.*?)(?=\n[A-Z][a-z]+:|\Z)"  # Match until next capitalized field or end
        match = re.search(pattern, text, re.DOTALL)
        if match:
            fields[key] = match.group(1).strip()
    return fields

# Parse all listings
parsed_listings = [parse_listing(listing) for listing in raw_listings]


# Pretty print the first parsed listing
import json
print(json.dumps(parsed_listings[0], indent=2))
```

## Step 3: Store listings in a vector database (ChromaDB)

In this step, we'll:

- Convert each listing into an embedding using OpenAI Embeddings.
- Store the embeddings along with metadata into ChromaDB for semantic search later.


### Step 3.1: Setup ChromaDB & OpenAI Embeddings

```python
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import Chroma

# Reuse OpenAI credentials already set in the environment
embedding_model = OpenAIEmbeddings()

# Prepare documents and metadata
texts = [listing["Description"] + " " + listing["Neighborhood Description"] for listing in parsed_listings]
metadatas = parsed_listings  # Attach the full structured info as metadata

# Create ChromaDB store
vectorstore = Chroma.from_texts(texts=texts, embedding=embedding_model, metadatas=metadatas, collection_name="homematch")
```

### Step 3.2: Test ChromaDB by retrieving one result

```python
# Try a test search
query = "A quiet neighborhood with eco-friendly homes"
results = vectorstore.similarity_search(query, k=1)

# Show result
import json
print(json.dumps(results[0].metadata, indent=2))
```

## Step 4: Collect buyer preferences
In this step, we’ll simulate how a buyer expresses their needs and wants — using either:
- Natural language input
- Structured questions


### Step 4: Define Buyer Preferences

```python
# Sample questions (optional — for context)
questions = [   
    "How big do you want your house to be?", 
    "What are 3 most important things for you in choosing this property?", 
    "Which amenities would you like?", 
    "Which transportation options are important to you?",
    "How urban do you want your neighborhood to be?",   
]

# Simulated buyer answers (natural language)
answers = [
    "A comfortable three-bedroom house with a spacious kitchen and a cozy living room.",
    "A quiet neighborhood, good local schools, and convenient shopping options.",
    "A backyard for gardening, a two-car garage, and a modern, energy-efficient heating system.",
    "Easy access to a reliable bus line, proximity to a major highway, and bike-friendly roads.",
    "A balance between suburban tranquility and access to urban amenities like restaurants and theaters."
]

# Combine into a single preference statement
buyer_profile = " ".join(answers)

# Preview the profile
print(buyer_profile)
```

## Step 5: Semantic search based on buyer preferences

We’ll use the buyer_profile as a semantic query against the vector database, and retrieve the top matching listings.

### Step 5: Perform semantic search

```python
# Perform the search with buyer preferences
top_k = 3  # Number of listings to return
search_results = vectorstore.similarity_search(buyer_profile, k=top_k)

# Display results
for i, result in enumerate(search_results):
    print(f"\n🏡 Match #{i+1}")
    print(json.dumps(result.metadata, indent=2))
```

# Step 6: Personalize the listing descriptions with GPT

We’ll ask the LLM to rewrite each listing's description in a way that subtly emphasizes what the buyer values (without changing facts).

### Step 6: Personalization prompt & generator


```python
# Initialize the LLM again if needed
llm = OpenAI(model_name="gpt-3.5-turbo-instruct", temperature=0.7)

# Function to personalize a listing
def personalize_listing(original_listing, buyer_profile):
    prompt = f"""
You are a helpful real estate assistant.

Here is a buyer's profile describing their preferences:
{buyer_profile}

Now, here is a real estate listing:
{json.dumps(original_listing, indent=2)}

Rewrite the listing's "Description" and "Neighborhood Description" to appeal to the buyer's preferences. 
Do not change factual details (e.g., price, number of rooms), just highlight what's most relevant to this buyer.

Output in this format:
Personalized Description:
[...]
Personalized Neighborhood Description:
[...]
    """
    return llm(prompt)

# Generate and display personalized descriptions
for i, result in enumerate(search_results):
    print(f"\n🎯 Personalized Listing #{i+1}")
    personalized_text = personalize_listing(result.metadata, buyer_profile)
    print(personalized_text.strip())
```
